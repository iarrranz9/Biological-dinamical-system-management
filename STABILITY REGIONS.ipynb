```python
from sympy import dsolve, Eq, symbols, Function, solve, diff, lambdify, log, sqrt
from sympy.plotting import plot
from matplotlib import pyplot as plt
import numpy as np
from math import e

def flowplt(l, ran=2.5):
    '''plot the system's flow 
    '''
    tvals=np.arange(-ran, ran, 0.1)
    xvals=l(tvals)
    
    plt.figure
    plt.plot(tvals, xvals)
    plt.xlabel('x')
    plt.ylabel('difx')
    ax=plt.gca()
    ax.spines['right'].set_color('none')
    ax.spines['top'].set_color('none')
    ax.xaxis.set_ticks_position('bottom')
    ax.spines['bottom'].set_position(('data', 0))
    ax.yaxis.set_ticks_position('left')
    ax.spines['left'].set_position(('data', 0))
    ax.yaxis.set_label_coords(-0.1, 0.5)
    ax.xaxis.set_label_coords(0.5, -0.1)
   
    return plt.show()


def an_fixed(f_points, flam):
    
    f_points_e=[]
    for n in f_points:
        if n.is_real==True:
            f_points_e.append(n)
    for i in f_points_e:
        if(flam(i-0.0001).is_positive) and (flam(i+0.0001).is_positive==False):
            print('stable:', i)
        elif(flam(i-0.0001).is_positive==False) and (flam(i+0.0001).is_positive):
            print('unstable:', round(i,3))
        else:
            print('halfstable:', round(i,3))

x=symbols('x')
r=symbols('r')
f=r*x+x**3/(1+x**2)

f_points=solve(f, x)
print(f_points)
print('\n')
print('3 intervals of interest')
print('\n')
print('Flow for r>0')

r1=4
flam1=lambdify(x, f.subs({r:r1}), 'numpy')
flowplt(flam1)
f_points1=solve(f.subs({r:r1}), x)
an_fixed(f_points1, flam1)

print('\n')
print('Flow for -1<r<0')
r2=-0.5
flam2=lambdify(x, f.subs({r:r2}), 'numpy')
flowplt(flam2)
f_points2=solve(f.subs({r:r2}), x)
an_fixed(f_points2, flam2)

print('\n')
print('Flow for r<0')
r3=-2
flam3=lambdify(x, f.subs({r:r3}), 'numpy')
flowplt(flam3)
f_points3=solve(f.subs({r:r3}), x)
an_fixed(f_points3, flam3)
```
