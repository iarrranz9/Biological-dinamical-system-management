### EXERCISE 1 

```python
from sympy import dsolve, Eq, symbols, Function, nonlinsolve, solve, diff, lambdify, log, sqrt, hessian, Matrix
from sympy.plotting import plot
from scipy.integrate import odeint
from matplotlib import pyplot as plt
import numpy as np


x = symbols('x')
y = symbols('y')


difx = -x + 1*y +(x**2)*y
dify = 2 - 1*y - (x**2)*y
fxp = nonlinsolve([difx, dify], [x, y])
fxp
```

```python
def model(X, t):
    return np.array([-X[0] + 1*X[1] + (X[0]**2)*X[1],
                    2 - 1*X[1] - (X[0]**2)*X[1]])


ts = np.linspace(0, 50, 1000)
x0 = [0.5, 0.5]
x1 = [0.6, 1.5]

xs = odeint(model, x0, ts)
xr = odeint(model, x1, ts)

plt.plot(xs[:, 0], xs[:, 1], 'black')
plt.plot(xr[:, 0], xr[:, 1], 'red')

plt.ylabel('diffy')
plt.xlabel('diffx')
plt.show()
```

```python

def model(X, t):
    return np.array([-X[0] + 1*X[1] + (X[0]**2)*X[1],
                    2 - 1*X[1] - (X[0]**2)*X[1]])

x = symbols('x')
y = symbols('y')

ts = np.linspace(0, 3, 50)
ic = np.linspace(0, 2, 5)
for r in ic: 
    for s in ic:
        x0 = [r, s]
        xs = odeint(model, x0, ts)
        plt.plot(xs[:, 0], xs[:, 1], 'r-',alpha=0.2)

plt.xlabel('x')
plt.ylabel('y')
plt.tick_params()

X, Y = np.mgrid[0:3.5:10j, 0:2.5:10j]
u = -X + Y - (X**2)*Y
v = 2 - Y - (X**2)*Y
plt.quiver(X, Y, u, v, color='b')

plt.show()
```
