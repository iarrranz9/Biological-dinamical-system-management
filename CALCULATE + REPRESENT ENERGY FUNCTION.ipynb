## 1 CALCULATE + REPRESENT ENERGY FUNCTION 

```python
#1) 
#To calculate the energy of the system, we multpily by the first derivative
from sympy import dsolve, Eq, symbols, Function, solve, diff, cos
import numpy as np
from matplotlib import pyplot as plt

#We set the values of x and v according to the exercise
x_values = np.linspace(-10,10,100)
v_values = np.linspace(-4,4,100)

#Creates X and Y matrices
X,V = np.meshgrid(x_values, v_values) 

#Evaluate the energy funcion at those values:
E = 1/2*(V**2) - np.cos(X) 

fig, ax = plt.subplots()
contour_diagram = ax.contour(X, V, E)
plt.xlabel(' θ', loc= 'right')
plt.ylabel(' v', loc='bottom')
plt.title('Energy countours for different values of E')
ax.spines['bottom'].set_position(('data', 0))
ax.spines['left'].set_position(('data', 0))
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')

plt.show()
```

## 2 ENERGY FUNCTION 3D

```python
import numpy as np 
import matplotlib.pyplot as plt

import mpl_toolkits

from mpl_toolkits.mplot3d import Axes3D
def fun(x,y):
    return 0.5*x**2-5*np.cos(y)
fig = plt.figure()
ax = fig.add_subplot(111,projection = '3d')
x = np.arange(-10.0,10.0,0.1)
y = np.arange(-4.0,4.0,0.1)
X, Y = np.meshgrid(x,y)
zs = np.array([fun(x,y) for x,y in zip(np.ravel(X),np.ravel(Y))])
Z = zs.reshape(X.shape)
ax.plot_surface(X,Y,Z)
ax.set_xlabel('theta')
ax.set_ylabel('phi')
ax.set_zlabel('theta')
plt.tick_params()
ax.view_init(30,-70)
plt.show()
```

## 3 CALCULATE + REPRESENT ENERGY FUNCTION 2D,3D

```python
import numpy as np
from matplotlib import pyplot as plt

x=np.linspace(-10,10,100)
y=np.linspace(-4,4,100)


X,Y = np.meshgrid(x, y) 

E = ((1/2)*(Y**2)-(1/2)*(X**2)+(1/4)*(X**4))

fig, ax = plt.subplots()
contour_diagram = ax.contour(X, V, E)
plt.xlabel('θ', loc= 'right')
plt.ylabel('v', loc='top')
plt.title('Energy countours for different values of E')
ax.spines['bottom'].set_position(('data', 0))
ax.spines['left'].set_position(('data', 0))
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
plt.show()
```

## 4 STUDY OF DAMPED SYSTEM, FX POINTS + PATHS

```python
#Calculating and classifying fixed points

import numpy as np

# Coeficients del sist lineal dX = JX, [dx1;dx2] = [a,b;c,d][x1;x2]
a = 0
b = 1
c = -5
d = -0.1

J = np.matrix([[a,b],[c,d]])
dJ = np.linalg.det(J)
tJ = np.trace(J)
print('The determinant of the Jacobian is = ',round(dJ,2),'; The trace of the Jacobian is = ',tJ)
print('----------------------------------------------------------------------------------------------')
if dJ < 0:
    print('Saddle point')
elif dJ == 0:
    print('Multiple saddle points')
else:
    if tJ == 0:
        print('Center')
    elif tJ > 0:
        stb = 'Unstable'
    else:
        stb = 'Stable'
    k = tJ**2-4*dJ
    if k < 0:
        print('Spiral '+ stb)
    elif k == 0:
        print('Star, Degenerated node '+ stb)
    else:
        print('Node '+ stb)  
```

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.integrate import odeint
import pylab as pl

# 2 dimensional linear syst
a,b,c,d = 0, 1, -5, -0.1

def dx_dt(x,t):
    return [a*x[0] + b*x[1],c*x[0] + d*x[1]]

# Trajectories in forward time
ts = np.linspace(0, 4, 100)
ic = np.linspace(-1, 1, 5)
for r in ic:
    for s in ic:
        x0 = [r,s]
        xs = odeint(dx_dt, x0, ts)
        plt.plot(xs[: , 0],xs[: , 1],'r--')

# Trajectories in backward time
ts = np.linspace(0, -4, 100)
ic = np.linspace(-1, 1, 5)
for r in ic:
    for s in ic:
        x0 = [r,s]
        xs = odeint(dx_dt, x0, ts)
        plt.plot(xs[: , 0],xs[: , 1],'r--')

plt.xlabel('x')
plt.ylabel('y')        
plt.tick_params()
plt.xlim(-1 , 1)
plt.ylim(-1 , 1)

X,Y = np.mgrid[-1: 1: 10j, -1: 1: 10j]
u = a*X + b*Y
v = c*X + d*Y
pl.quiver(X, Y, u, v, color='b')
plt.show()
```
